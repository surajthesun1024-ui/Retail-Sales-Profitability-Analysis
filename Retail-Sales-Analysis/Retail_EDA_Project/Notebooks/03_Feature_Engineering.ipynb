```python
from google.colab import drive
drive.mount('/content/drive')

project_path="/content/drive/MyDrive/Retail_EDA_Project/"

import pandas as pd

df= pd.read_csv(project_path+"Data/Processed/cleaned_superstore.csv")
```

```python
df.head()
```

```python
#Profit Margin Feature
df['Profit Margin'] = (df['Profit']/df['Sales'])*100
```

```python
#Discount Category
def discount_category(x):
  if x==0:
    return 'No Discount'
  elif x <= 0.2:
    return 'Low Discount'
  elif x <=0.4:
    return 'Medium discount'
  else:
    return 'High Discount'

df['Discount Category']= df['Discount'].apply(discount_category)
```

```python
#Loss Flag
df['Loss Flag']= df['Profit'].apply(
    lambda x: 'Loss' if x<0 else 'Profit'
)
```
