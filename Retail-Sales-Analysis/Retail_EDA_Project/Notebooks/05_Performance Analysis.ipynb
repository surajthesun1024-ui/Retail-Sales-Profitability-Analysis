```python
from google.colab import drive
drive.mount('/content/drive')

project_path= "/content/drive/MyDrive/Retail_EDA_Project/"

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df= pd.read_csv(project_path+"Data/Processed/cleaned_superstore.csv")
```

```python
df.head()
```

```python
#Analysis of Sales & Profit by category

category_sales= df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

```python
category_sales
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=category_sales.index, y=category_sales.values, palette='Paired', hue=category_sales.index)

plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.title('Total Sales by Category')

plt.show()
```

```python
category_profit= df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
category_profit
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=category_profit.index, y=category_profit.values, palette='Paired', hue=category_profit.index)

plt.xlabel('Category')
plt.ylabel('Total Profit')
plt.title('Total Profit by Category')

plt.show()
```
