```python
from google.colab import drive
drive.mount('/content/drive')

project_path= "/content/drive/MyDrive/Retail_EDA_Project/"

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df= pd.read_csv(project_path+"Data/Processed/cleaned_superstore.csv")
```

```python
df.head()
```

```python
# Discount vs Profit Analysis
plt.figure(figsize=(6,4))
sns.scatterplot(x='Discount', y='Profit', data=df, palette='mako', hue='Discount', legend=False)

plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Discount vs Profit')

plt.show()
```

```python
# Discount vs Profit margin Analysis
plt.figure(figsize=(6,4))
sns.scatterplot(x='Discount', y='Profit Margin', data=df, palette='mako', hue='Discount', legend=False)

plt.xlabel('Discount')
plt.ylabel('Profit Margin')
plt.title('Discount vs Profit Margin')

plt.show()
```

```python
# Sales vs Profit Analysis
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales', y='Profit', data=df, palette='rocket', hue='Sales', legend=False)

plt.title('Sales vs Profit Analysis')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
# Sales vs Profit margin analysis
plt.figure(figsize=(6,4))
sns.scatterplot(x='Sales', y='Profit Margin', data=df, palette='rocket', hue='Sales', legend=False)

plt.xlabel('Sales')
plt.ylabel('Profit Margin')
plt.title('Sales vs Profit Margin')

plt.show()
```

```python
# Discount Category vs Profit Analysis
plt.figure(figsize=(6,4))
sns.violinplot(x='Discount Category', y='Profit', data=df, palette='magma', hue='Discount Category', legend=False)

plt.xlabel('Discount Category')
plt.ylabel('Profit')
plt.title('Average Profit by Discount Category')

plt.show()
```

```python
# Discount category vs Profit margin Analysis
plt.figure(figsize=(6,4))
sns.violinplot(x='Discount Category', y='Profit Margin', data=df, palette='magma', hue='Discount Category', legend=False)

plt.xlabel('Discount Category')
plt.ylabel('Profit Margin')
plt.title('Average Profit Margin by Discount Category')

plt.show()
```

```python
# Category vs Profit Margin Analysis
plt.figure(figsize=(6,4))
sns.barplot(x='Category', y='Profit Margin', data=df, palette='dark', hue='Category', legend=False)

plt.xlabel('Category')
plt.ylabel('Profit Margin')
plt.title('Average Profit Margin by Category')

plt.show()
```

```python
# Segment vs Profit Margin Analysis
plt.figure(figsize=(6,4))
sns.barplot(x='Segment', y='Profit Margin', data=df, palette='Greens', hue='Segment', legend=False)

plt.xlabel('Segment')
plt.ylabel('Profit Margin')
plt.title('Average Profit Margin by Segment')

plt.show()
```

```python
# Region vs Profit Margin
plt.figure(figsize=(6,4))
sns.barplot(x='Region', y='Profit Margin', data=df, palette='pastel', hue='Region', legend=False)

plt.xlabel('Region')
plt.ylabel('Profit Margin')
plt.title('Average Profit Margin by Region')

plt.show()
```

```python
# Correlation Heatmap (To find relationship between Sales, Profit, Profit margin, Discount, Quantity)
plt.figure(figsize=(8,4))

sns.heatmap(df[['Sales','Profit','Profit Margin','Discount','Quantity']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')

plt.show()
```
